### Japanese Investors are presently staying at The Manila Hotel so we will find the exact location of the hotel in order for us to locate the different restaurants in a radius of 2000 meters to see how popular the japanese restaurants in Manila

```python
address = 'One Rizal Park, Manila, Philippines'

geolocator = Nominatim(user_agent="foursquare_agent")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(latitude, longitude)
```

## We will search a Japanese restaurant in Metro manila within a 200 radius.

```python
search_query = 'japanese restaurant'
radius = 2000
print(search_query + ' .... OK!')
url = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&oauth_token={}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude,ACCESS_TOKEN, VERSION, search_query, radius, LIMIT)
results = requests.get(url).json()
# assign relevant part of JSON to venues
venues = results['response']['venues']
# tranform venues into a dataframe
dataframe = json_normalize(venues)
dataframe.head()
```

```python
dataframe.shape
```

```python
jap_dataframe = dataframe
```

### As we have seen in the dataframe there are 29 restaurants located in Manila and 75% of it is dominated by japanese cuisine.

```python
# keep only columns that include venue name, and anything that is associated with location
filtered_columns = ['name', 'categories'] + [col for col in jap_dataframe.columns if col.startswith('location.')] + ['id']
jap_dataframe_filtered = jap_dataframe.loc[:, filtered_columns]

# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

# filter the category for each row
jap_dataframe_filtered['categories'] = jap_dataframe_filtered.apply(get_category_type, axis=1)

# clean column names by keeping only last term
jap_dataframe_filtered.columns = [column.split('.')[-1] for column in jap_dataframe_filtered.columns]

jap_dataframe_filtered
```

### Let's visualize the restaurants that are located at the Heart of Manila

```python
jap_map = folium.Map(location=[latitude, longitude], zoom_start=13) # generate map centred around the Manila Hotel

# add a red circle marker to represent the Manila Hotel
folium.CircleMarker(
    [latitude, longitude],
    radius=10,
    color='red',
    popup='Conrad Hotel',
    fill = True,
    fill_color = 'red',
    fill_opacity = 0.6
).add_to(jap_map)

# add the japanese restaurants as blue circle markers
for lat, lng, label in zip(jap_dataframe_filtered.lat, jap_dataframe_filtered.lng, jap_dataframe_filtered.categories):
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        color='blue',
        popup=label,
        fill = True,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(jap_map)

# display map
jap_map
```
